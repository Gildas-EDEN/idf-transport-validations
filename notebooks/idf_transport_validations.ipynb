```python
# Clean the 'NB_VALIDATIONS' column: remove spaces (thousands separators) and convert the values to numeric.
df_all['NB_VALIDATIONS'] = df_all['NB_VALIDATIONS'].astype(str).str.replace(' ', '', regex=False)
df_all['NB_VALIDATIONS'] = pd.to_numeric(df_all['NB_VALIDATIONS'], errors='coerce')

# Drop the rows where the 'NB_VALIDATIONS' column became NaN after conversion.
df_all.dropna(subset=['NB_VALIDATIONS'], inplace=True)

# Filter non-negative values then convert to integers
df_all = df_all[df_all["NB_VALIDATIONS"] >= 0].copy()
df_all["NB_VALIDATIONS"] = df_all["NB_VALIDATIONS"].astype(int)
```

```python
df_all["NB_VALIDATIONS"].describe()
```

### Explanation for the number of validations

The NB_VALIDATIONS field shows an atypical distribution, characterized by a high proportion of values that are multiples of 5 and by the near-total absence of very low isolated values. This particularity is explained by the anonymization process implemented by Ile-de-France Mobilites: low validation volumes are not provided exactly for confidentiality reasons. Thus, a value equal to 5 corresponds to 5 validations or fewer, and more generally small values are rounded or grouped.

### Strategy to address this encoding choice

* Keep the values as they are for all aggregated analyses (by station, by day, by period): the rounding introduces a very small bias at large scale.

* Avoid interpreting the values at an ultra-fine level (for example, "exactly 5 validations" as an exact figure) and consider them as approximate values.

* We may optionally force the integer type and replace negative or aberrant values if any are found.

## Analysis of the impact of the 2024 Summer Olympics

Graphically highlight which stops were the most impacted by the Olympics.

Use all ticket categories.

Explain your method.

**Reminder**: the Olympics took place from 26 July 2024 to 11 August 2024.



```python
## Your code to compare the traffic at a stop between the 2024 Olympics period and the comparison period

# Import the datetime class to manipulate dates
from datetime import datetime

# Define the Olympics start date (26 July 2024)
debut_jo = datetime(2024, 7, 26)
# Define the Olympics end date (11 August 2024)
fin_jo = datetime(2024, 8, 11)

# Filter the data for the dates within the Olympics period
df_jo = df_all[(df_all["DATE"] >= debut_jo) & (df_all["DATE"] <= fin_jo)]

# Filter the data for the reference period (21 days before the Olympics)
df_ref = df_all[(df_all["DATE"] >= debut_jo - pd.Timedelta(days=21)) &
                (df_all["DATE"] < debut_jo)]

# Compute the total traffic per stop during the Olympics
trafic_jo = df_jo.groupby("NOM_ARRET")["NB_VALIDATIONS"].sum()

# Compute the total traffic per stop over the reference period
trafic_ref = df_ref.groupby("NOM_ARRET")["NB_VALIDATIONS"].sum()

# Create a comparison DataFrame between Olympics traffic and reference traffic
impact = pd.DataFrame({
    "JO": trafic_jo,
    "Reference": trafic_ref
}).dropna()  # Remove stops with no data in either of the two periods

# Compute the percentage change between the two periods
impact["Variation_%"] = 100 * (impact["JO"] - impact["Reference"]) / impact["Reference"]

# Select the 20 stops with the largest traffic increase
top20 = impact.sort_values("Variation_%", ascending=False).head(20)

# Display the top 20
top20
```

```python
## Graphical visualization of the stations most impacted by the 2024 Olympics

plt.figure(figsize=(12,6))
plt.barh(top20.index, top20["Variation_%"])
plt.title("Stations most impacted by the 2024 Olympics")
plt.xlabel("Variation (%)")
plt.gca().invert_yaxis()
plt.show()
```

## Modeling with Scikit-Learn

We try to build a basic model to forecast traffic for the next 7 days, for one station.

Scikit-Learn does not include a model for time series.

The classic approach is then to use a standard regression technique (linear regression of course, but also RandomForestRegressor for example).

For a first simple version, use a set of input variables (features) made of the validations over the previous 14 days (X) to predict the validations of the day (y). In this version, we will use either a `LinearRegression` or a `RandomForestRegressor`.

The code must include:
- the preparation of the _features_ (X)
- the construction of a training and test set
- the training of a model
- the measurement of the model performance: RMSE

Reminder: work only on the two main ticket categories.


```python
# Your code here

## Data preparation (Gare de Lyon is the chosen station)

serie = trafic_station("GARE DE LYON")

df_ml = pd.DataFrame({"y": serie})

for i in range(1, 15):
    df_ml[f"lag_{i}"] = df_ml["y"].shift(i)

df_ml = df_ml.dropna()
```

```python
## Split train/test

from sklearn.model_selection import train_test_split

X = df_ml.drop(columns=["y"])
y = df_ml["y"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
## Model training

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=200, random_state=0)
model.fit(X_train, y_train)
```

```python
## Performance (RMSE)

from sklearn.metrics import mean_squared_error
import numpy as np

pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, pred))
rmse
```

```python
##  Display the actual and predicted values in a table

# Table comparing actual and predicted values
resultats = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": pred
}, index=y_test.index)

resultats.head(20)   # display the first 20 rows
```

```python
## Display the estimates as a chart

plt.figure(figsize=(14,5))
plt.plot(y_test.index, y_test.values, label="Actual", linewidth=2)
plt.plot(y_test.index, pred, label="Predicted", linewidth=2)
plt.title("Comparison of actual and predicted values")
plt.xlabel("Date")
plt.ylabel("Validations")
plt.legend()
plt.grid(True)
plt.show()
```

```python
## Display the errors day by day

resultats["Error"] = resultats["Actual"] - resultats["Predicted"]
resultats["Abs_Error"] = resultats["Error"].abs()

resultats.head(20)
```

```python
##  Display the 10 best-predicted days and the 10 worst-predicted days

# Best-predicted days
print("Best-predicted days:")
display(resultats.sort_values("Abs_Error").head(10))

# Worst-predicted days
print("Worst-predicted days:")
display(resultats.sort_values("Abs_Error", ascending=False).head(10))
```

## Improving the model

Discuss ways to improve this first version of the model.


* Improving the model starts with enriching the information used for prediction. Indeed, transport traffic depends heavily on external factors such as the day of the week, school holidays, public holidays, the weather, or exceptional events. Adding these explanatory variables would allow the model to better understand the natural variations in traffic. We can also improve the temporal representation by increasing the historical depth: using 30 or 60 days of past data, or moving averages, helps capture longer trends and smooth out one-off fluctuations. Finally, smoothing the data beforehand helps reduce daily noise and makes the model easier to train.

* Another way to improve the model is to use models specifically designed for time series. Statistical approaches such as ARIMA or SARIMA, or more advanced models such as Prophet or recurrent neural networks (LSTM/GRU), are able to better exploit the temporal structure of the data. In parallel, hyperparameter optimization through techniques such as GridSearchCV or RandomizedSearchCV would allow fine-tuning the model behavior for better performance. Together, these improvements would make the prediction more robust, more stable, and more accurate, accounting for both the temporal dynamics and the contextual factors influencing traffic.
